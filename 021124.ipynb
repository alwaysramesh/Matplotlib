```python
import numpy as np
import matplotlib.pyplot as plt

# First subplot
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 1)
plt.plot(x, y)
plt.title("Line Plot 1")

# Second subplot
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 2)
plt.plot(x, y)
plt.title("Line Plot 2")

# Third subplot
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 3)
plt.plot(x, y)
plt.title("Line Plot 3")

# Fourth subplot
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 4)
plt.bar(x, y)
plt.title("Bar Chart")

# Fifth subplot (Pie chart)
sizes = np.array([3, 8, 1, 10])  # Use meaningful sizes for the pie chart
plt.subplot(2, 3, 5)
plt.pie(sizes, labels=x)
plt.title("Pie Chart")

# Sixth subplot
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 6)
plt.plot(x, y)
plt.title("Line Plot 4")

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
# 2 row 2 column
fig,a=plt.subplots(2,2)
import numpy as np
x=np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('squre')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('squre root')

a[1][0].plot(x,np.exp(x))

a[1][0].set_title('exp')

a[1][1].plot(x,np.log10(x))

a[1][1].set_title('log')
plt.show()
```
